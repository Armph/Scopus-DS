```python
import os
import csv

directory = "2023"

# Loop through every file in the directory
for filename in os.listdir(directory):
    # Check if the entry is a file
    if os.path.isfile(os.path.join(directory, filename)):
        path = os.path.join(directory, filename).replace('\\', '/')
        print(path)
        with open(path, "r",encoding='utf-8') as f:
            json_content = f.read()
        json_data = json.loads(json_content)
        class_list = json_data["abstracts-retrieval-response"]["item"]["bibrecord"]["head"]["enhancement"]["classificationgroup"]["classifications"]
        for c in class_list:
            if c["@type"] == "SUBJABBR":
                if isinstance(c["classification"], str):
                    subject_areas =  c["classification"]
                else:
                    subject_areas = []
                    for k in c["classification"]:
                        subject_areas.append(k["$"])
                    subject_areas = ",".join(subject_areas)
        eid = json_data["abstracts-retrieval-response"]["coredata"]["eid"]
        publication_year = json_data["abstracts-retrieval-response"]["item"]["bibrecord"]["head"]["source"]["publicationyear"]['@first']
        # print(json_data["abstracts-retrieval-response"]["item"]["bibrecord"]["head"]["abstracts"])
        abstract = json_data["abstracts-retrieval-response"]["item"]["bibrecord"]["head"]["abstracts"]
        if abstract != None:
            length_of_abstract = len(json_data["abstracts-retrieval-response"]["item"]["bibrecord"]["head"]["abstracts"].strip().split())
        else:
            length_of_abstract = 0
        author_keywords_list = json_data["abstracts-retrieval-response"]["authkeywords"]
        if author_keywords_list != None and isinstance(author_keywords_list["author-keyword"], list):
            author_keywords = []
            for c in author_keywords_list["author-keyword"]:
                author_keywords.append(c["$"])
            author_keywords = ",".join(author_keywords)
        elif author_keywords_list != None and isinstance(author_keywords_list["author-keyword"], dict):
            author_keywords = [author_keywords_list["author-keyword"]['$']]
            author_keywords = ",".join(author_keywords)
        else:
            author_keywords = None
        sub_type = json_data["abstracts-retrieval-response"]["coredata"]["subtype"]
        try:
            citation_count = int(json_data["abstracts-retrieval-response"]["coredata"]["citedby-count"])
        except (TypeError, IndexError):
            citation_count = None
        try:
            refcount = int(json_data["abstracts-retrieval-response"]["item"]["bibrecord"]["tail"]["bibliography"]["@refcount"])
        except (TypeError, IndexError):
            refcount = None
        data = [eid,subject_areas,publication_year,length_of_abstract,author_keywords,sub_type,citation_count,refcount]
        with open("2023.csv","a",newline='',encoding="utf-8") as csv_file:
            writer = csv.writer(csv_file)
            writer.writerow(data)
```
